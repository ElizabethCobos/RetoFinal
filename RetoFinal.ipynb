#Aplicar método de distribución normal para identificar Outliers, tomando como limites 2 desviaciones estándar, exclusivamente para las columnas “gnp old” y “population”, obtener los diagramas de caja de cada una respectivamente.

```python
#GNP_OLD
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data7["gnp_old"]
Limite_Superior= y.mean() + 2*y.std()
Limite_Inferior= y.mean() - 2*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers
outliers= data7[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos limpios
data_clean= data7[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data_clean
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean["gnp_old"]) 
plt.title("Outliers de gnp_old")
plt.show() #dibujamos el diagrama
```

```python
#POPULATION
#Método aplicando desviación estandar. Encuentro los valores extremos
y=data7["population"]
Limite_Superior= y.mean() + 2*y.std()
Limite_Inferior= y.mean() - 2*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
#Encontramos Ouliers
outliers= data7[(y>Limite_Superior)|(y<Limite_Inferior)]
outliers
```

```python
#Obtenemos datos limpios
data_clean= data7[(y<=Limite_Superior)&(y>=Limite_Inferior)]
data_clean
```

```python
#Realizamos diagrama de caja o bigote
fig = plt.figure(figsize =(5, 3))
plt.boxplot(data_clean["population"]) 
plt.title("Outliers de population")
plt.show() #dibujamos el diagrama
```

```python
import pandas as pd
import numpy as np
import cufflinks as cf
import matplotlib.pyplot as plt
from IPython.display import display,HTML

cf.set_config_file(sharing='public',theme='white',offline=True) # write cf.getThemes() to find all themes available
```

```python
#Reestructurar el Dataframe
#(index=‘code', columns=‘continent',  values=‘population')
data8 = data8.dropna()
Data_res1= data8.pivot(index='code', columns='continent',  values='population')
Data_res1
```

```python
#Lineplot o grafico lineal
Data_res1.iplot(kind='line', xTitle='code', yTitle='population', title='population by continent')
```

```python
# Barplot o grafico de barras
population_muestra = Data_res1[Data_res1.index.isin(["HuN", "NZL"])]

# plotting
population_muestra.iplot(kind='bar', xTitle='code',
                           yTitle='continent', title='Muestra')
```

```python
# Boxplot o grafico de caja

Data_res1.iplot(kind='box', xTitle='Continent',
                    yTitle='Population', title='Población por continente')
```

```python
#Histplot o Histograma

Data_res1[['Asia']].iplot(kind='hist',xTitle='Population', 
                               title='Distibución de la poblacion en Asia')
```

```python
#Scatterplot o Mapa de dispersión

Data_res1[['Asia']].iplot(kind='scatter', mode='markers',xTitle='code', 
                               title='Distibución de la población en Asia')
```
